Basic charts and visual encoding

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep_Hours': [6,7,6,8,7],
    'Assignment_Completed':[1,2,1,3,2]
}
df = pd.DataFrame(data)
df
```

```python
# Line chart
plt.plot(df['Days'], df['Study_Hours'],marker='*')
plt.title('Prince Kumar Sharma')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
plt.figure(figsize=(7,4))

plt.plot(df['Days'], df['Study_Hours'],marker='o', color='blue',label='Study Hours')
plt.plot(df['Days'], df['Marks'],marker='s', color='red',label='Marks')
plt.title('Prince Kumar Sharma')
plt.xlabel('Days')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Bar chart
plt.bar(df['Days'], df['Marks'],color='green')
plt.title('Prince Kumar Sharma')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
bars = plt.bar(df['Days'], df['Marks'],color='red')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y), ha='center', va='bottom')
plt.title('Prince Kumar Sharma')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y')
plt.show()
```

```python
# Histogram
data1 = {
    'Marks':[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
}
df = pd.DataFrame(data)
plt.hist(df['Marks'], bins=3,edgecolor='black',alpha=0.300)
plt.title('Prince Kumar Sharma')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
marks = [50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]

mean_value = np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks, bins=6,color='skyblue',edgecolor='black')

plt.axvline(mean_value, color='red', linestyle='--', linewidth=2,label='Mean')
plt.legend()
plt.title('Prince Kumar Sharma')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```
